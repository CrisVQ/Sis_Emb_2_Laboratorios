```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers


xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
ys = 3 * xs + 2                                            #Y = 3X + 2


model = keras.Sequential()
model.add(layers.Dense(1, input_dim=1))                    #



model.compile(optimizer=keras.optimizers.Adam(0.01),
              loss="mean_squared_error",
              metrics=["mean_squared_error"])


model.fit(xs, ys, epochs=1000, verbose=0)



x_test = np.array([5, 3.3], dtype=float)
predictions = model.predict(x_test)

print("Predicción para X = 5:", predictions[0])
print("Predicción para X = 3.3:", predictions[1])


loss, mse = model.evaluate(xs, ys)
print("Pérdida (MSE) en los datos de entrenamiento:", mse)
```
